```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = 'https://raw.githubusercontent.com/S1mDu/LCA/main/Data/'
```

```python
# Update the filepath to the local directory where the files are stored

demo = pd.read_excel(filepath + 'demography.xlsx')
#demo = pd.read_excel(demo_url)
tot = pd.read_excel(filepath + 'DR1TOT.xlsx')
```

```python
df = pd.merge(demo, tot, how='outer', on='SEQN')
df.head()
```

```python
df.shape
```

```python
df['INDFMPIR']
```

```python
df[df['INDFMPIR'] < 1.2].mean()
```

```python
macronutrient = ['DR1TKCAL', 'DR1TPROT', 'DR1TCARB', 'DR1TSUGR', 'DR1TFIBE', 'DR1TTFAT', 'DR1TSFAT', 'DR1TMFAT', 'DR1TPFAT']
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['INDFMPIR'], bins=5, color='blue', alpha=0.7)
plt.title('Distribution of INDFMPIR')
plt.xlabel('INDFMPIR')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.25)
plt.show()
```

```python
# Create 5 bins for INDFMPIR
bins = pd.cut(df['INDFMPIR'], bins=4, labels=False)
# Group by the bins and calculate the mean of DR1TKCAL
grouped = df.groupby(bins)['DR1TKCAL'].mean().reset_index()
# Rename the columns for clarity
grouped.columns = ['INDFMPIR_Bin', 'Mean_DR1TKCAL']
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(grouped['INDFMPIR_Bin'], grouped['Mean_DR1TKCAL'], color='blue', alpha=0.7)
plt.title('Mean KCalories by INDFMPIR Bin')
plt.xlabel('INDFMPIR Bin')
plt.ylabel('Mean DR1TKCAL')
plt.xticks(grouped['INDFMPIR_Bin'], [f'Bin {i}' for i in grouped['INDFMPIR_Bin']])
plt.grid(axis='y', alpha=0.25)
plt.show()
```
